```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data=pd.read_csv('titanic.csv')
```

```python
data=data.drop(["Name"],axis=1)
```

```python
data_survived = data.loc[data['Survived'] == 1]
data_notSurvived = data.loc[data['Survived'] == 0]
```

```python
data_notSurvived.rename(columns={"Survived": "Not survived"}, inplace=True)
data_notSurvived
```

```python
data_survivedFemale = data_survived.loc[data['Sex'] == "female"]
data_survivedmale = data_survived.loc[data['Sex'] == "male"]
data_notSurvivedFemale = data_notSurvived.loc[data['Sex'] == "female"]
data_notSurvivedmale = data_notSurvived.loc[data['Sex'] == "male"]
```

```python
plt.figure(figsize=(8, 8))
sns.histplot(data_survivedFemale['Sex'], bins=10, kde=True, color='pink')
sns.histplot(data_survivedmale['Sex'], bins=10, kde=True, color='blue')
plt.title('Distribution des survivants par Sex . ')
plt.xlabel('Sex')
plt.ylabel('Nombre de voyageur')
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.histplot(data_survivedFemale['Sex'], bins=20, kde=True, color='pink')
sns.histplot(data_survivedmale['Sex'], bins=20, kde=True, color='blue')
sns.histplot(data_notSurvivedFemale['Sex'], bins=20, kde=True, color='green')
sns.histplot(data_notSurvivedmale['Sex'], bins=20, kde=True, color='orange')
plt.title('Distribution des survivants et des non survivants par Sex . ')
plt.xlabel('Sex')
plt.ylabel('Nombre de voyageur')
plt.show()
```

```python
plt.hist(data_notSurvived['Pclass'])
plt.title('Histogramme des classes pour les non survivants')
plt.xlabel('Classe')
plt.ylabel('Nombre de non survivants')
plt.show()
```
